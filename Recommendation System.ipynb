```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
data=pd.read_csv(r"C:\Users\asima\Downloads\book.csv", encoding='latin-1')
data
```

```python
data1=data.drop(['Unnamed: 0'],axis=1)
data1
```

```python
data1.isnull().any()
```

```python
data1.describe().T
```

```python
data2=data1.rename({'User.ID':'userid','Book.Title':'bookstitle','Book.Rating':'booksrating'},axis=1)
data2
```

```python
data3=data2.pivot_table(index='userid',
                           columns='bookstitle',
                           values='booksrating').reset_index(drop=True)
data3
```

```python
data3.index=data2.userid.unique()
```

```python
data3.head(25)
```

```python
data3.fillna(0,inplace=True)
data3
```

```python
user_sim=1-pairwise_distances(data3.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index=data2.userid.unique()
user_sim_df.columns=data2.userid.unique()
```
